```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../data/Synthetic_app_data.csv")
df.describe()
```

```python
df.columns
```

```python
sns.pairplot(df)
```

```python
indep = df[["Number of Bugs Reported","Number of Features"]]
indep
dep = df.loc[:,"User Satisfaction Score"]
```

```python
train_x, test_x, train_y, test_y = train_test_split(indep, dep, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(train_x, train_y)
```

```python
y_hat = model.predict(test_x)
```

```python
mae = mean_absolute_error(y_hat, test_y)
mae
```

```python
model.intercept_
```

```python
model.coef_
```

```python
print(f"The Eval Metrix for MLR is {mae}")
```
