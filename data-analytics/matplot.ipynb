```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

years = np.array([2018, 2019, 2020, 2021, 2022])
publish = np.array([1500, 1750, 2100, 2300, 2500])
launch = np.array([50, 75, 110, 150, 200])

plt.xlabel("Year")
plt.ylabel("Count")
plt.title("AI Research and Applications Over Years")

plt.plot(years, publish, marker='o', linestyle='--', label="AI Research Papers")
plt.plot(years, launch, marker='o', label="AI Applications Launched")

plt.xticks(years)
plt.legend()
plt.show()
```
